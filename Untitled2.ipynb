```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import poisson
import pandas as pd
```

```python
df = pd.read_csv('cancer-probabilities.csv')
df.head()
```

```python
#binomial
n = 20
p = 0.6
x = np.arange(0, n+1)

bino = binom.pmf(x, n, p)
plt.figure(figsize=(10,6))
plt.plot(x, bino, color='blue')
plt.title("binomial distribution")
plt.show()
```

```python
rate = 2
x = np.arange(0,20)

poi = poisson.pmf(x, rate)

plt.figure(figsize=(10, 6))
plt.bar(x, poi, color='skyblue', alpha=0.7, label='Poisson PMF')
plt.title(f'Poisson Distribution ( = {rate})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability Mass Function (PMF)')
plt.xticks(x)
plt.legend()
plt.grid(axis='y')
```
