```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df = pd.read_csv('zomato-vs-swiggy-tweets.csv')
df.head()
```

```python
mean = df['likeCount'].mean()
sd = df['likeCount'].std()

z_score = (df['likeCount'].max() - mean) / sd

x = np.linspace(mean - 4*sd, mean + 4*sd, 1000)
y = norm.pdf(x, mean, sd)
z = mean + z_score * sd
print(z)
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='Normal Distribution', lw=2)
plt.axvline(z, color='red', linestyle='--', label='z-score')
plt.show()
```
